```python
df = pd.read_csv("C:\\Users\\ohrtt\\Documents\\GitHub\\c3_2024\\data_analysis\\select_target\\data.csv")
df
```

```python
df['hour_angle_at_19']=19*15-df['ra']
df['hour_angle_at_22']=22*15-df['ra']
```

```python
def calc_alt(lati, dec, H):
    return np.degrees(np.arcsin(np.sin(np.radians(dec))*np.sin(np.radians(lati))+np.cos(np.radians(dec))*np.cos(np.radians(lati))*np.cos(np.radians(H))))
```

```python
lati = 35.0 # 京都の緯度
df['alt_at_19'] = calc_alt(lati, df['dec'], df['hour_angle_at_19'])
df['alt_at_22'] = calc_alt(lati, df['dec'], df['hour_angle_at_22'])
```

```python
def calc_az(lati, dec, H, alt):
    sinA = -1.*np.cos(np.radians(dec))*np.cos(np.radians(lati))*np.sin(np.radians(H))/np.cos(np.radians(alt))
    cosA = (np.sin(np.radians(dec))-np.sin(np.radians(alt))*np.sin(np.radians(lati)))/np.cos(np.radians(alt))/np.cos(np.radians(lati))
    az = np.degrees(np.arctan2(sinA, cosA))
    az = np.where(az<0, az+360, az)
    return az
```

```python
df['az_at_19'] = calc_az(lati, df['dec'], df['hour_angle_at_19'], df['alt_at_19'])
df['az_at_22'] = calc_az(lati, df['dec'], df['hour_angle_at_22'], df['alt_at_22'])
```
