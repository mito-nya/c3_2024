```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import batman
```

```python
params = batman.TransitParams()
params.t0 = 1.                       #time of inferior conjunction
params.per = 2.5                     #orbital period
params.rp = 0.09                     #planet radius (in units of stellar radii)
params.a = 15.                       #semi-major axis (in units of stellar radii)
params.inc = 90.                     #orbital inclination (in degrees)
params.ecc = 0.                      #eccentricity
params.w = 90.                       #longitude of periastron (in degrees)
params.u = []                        #limb darkening coefficients [u1, u2]
params.limb_dark = "uniform"         #limb darkening model

params.fp = 3e-4
params.t_secondary = 2.25
```

```python
t = np.linspace(0, 4.5, 1000)
m = batman.TransitModel(params, t)    #initializes model
flux = m.light_curve(params)          #calculates light curve
```

```python
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1,1,1)
ax.plot(t, flux * 100, color="black", lw=3)
ax.plot(t, [99.19]*1000, color="black", ls="--", lw=1)
ax.set_xlim([0, 4.5])
ax.set_xlabel("時間 (日)", fontsize=16)
ax.set_ylabel("相対的な明るさ (%)", fontsize=16)
ax.set_yticks([99.19, 99.46, 99.73, 100])
ax.tick_params(labelsize=16)
plt.tight_layout()
plt.savefig("fig1.png")
plt.show()
```
